```python
#Usando el loc extrae solo las columnas de "Oro", "Plata" y "Bronce" de China e Irán.

df_medallas.loc[1:4:3][["Oro", "Plata","Bronce"]]
```

```python
# Usando el iloc extrae la información de la India. Devolved todas las columnas.

df_medallas.iloc[5]
```

```python
# Usando el iloc extrae todas las columnas de los países que estén en posición impar. La primera que nos tendría que salir es Japón y la última Corea del Norte.

df_medallas.iloc[1::2]
```

```python
# Utilizando el código del ejercicio anterior, selecciona unicamente las columnas "Oro", "Plata" y "Bronce".

df_medallas.iloc[1::2][["Oro", "Plata","Bronce"]]
```

```python
# ¿Qué país consiguió más de 200 medallas de oro? 

maskOro = df_medallas["Oro"] > 200

maskOro

df_medallas[maskOro].head(10)
```

```python
#¿Qué país consiguió más de 200 medallas de oro y más de 1000 de plata?

maskPlata = df_medallas["Plata"] > 1000

maskPlata

df_medallas[maskOro & maskPlata].head(10)
```

```python
#¿Qué país consiguió más de 200 medallas de oro o más de 700 de bronce?

maskBronce = df_medallas["Bronce"] > 700

maskBronce

df_medallas[maskOro | maskBronce].head(10)
```

```python
#Crea una nueva columna que se llame "total" que sea el resultado de la suma de todas las medallas obtenidas por cada país.

ColumTotal = df_medallas["Oro"] + df_medallas["Plata"] + df_medallas["Bronce"]

ColumTotal

df_medallas["Total_medallas"] = ColumTotal

df_medallas.head(10)
```

## SERIES

```python
'''Crea dos listas:

    Una que se llame "valores" que tenga 7 números aleatorios.

    Otra que se llame "indices" que tenga 7 strings que se llamen como los días de la semana.'''

valores = [random.randint(0, 50) for x in range(7)]

print(valores)

indices = ["Lunes","Martes","Miercoles","Jueves","Viernes","Sabado","Domingo"]

print(indices)
```

```python
#A partir de estas dos listas crea una Serie donde la variable "valores" sean los valores de la Serie y la variable indices sean los índices de la Serie.

semanaSerie = pd.Series(valores,index=indices)

display(semanaSerie)
```

```python
#A partir de las dos listas anteriores crea un diccionario usando un bucle for. Para hacer esto tenemos el método zip que nos permite iterar por dos listas a la vez. 
# Más información sobre este método https://www.programiz.com/python-programming/methods/built-in/zip.

result = zip(indices, valores)
semanaDiccionario = dict(result)


print(semanaDiccionario)

```

```python
#Crea una Serie con el diccionario creado en el punto anterior.

serie_semana = pd.Series(semanaDiccionario)

serie_semana.head()
```

```python
#¿Qué forma tiene esta última Serie?

serie_semana.shape
```

```python
#¿Qué tipo de datos tiene?

serie_semana.dtype
```

```python
#¿Qué tamaño tiene?

serie_semana.size
```

```python
#¿Cuáles son los índices de la Serie?

serie_semana.index
```

```python
#¿Y los values?

serie_semana.values
```

# Más ejercicios de loc e iloc

```python
worldData = pd.read_csv("world-data-2023_part1.csv",index_col=0)

worldData.head()
```

```python
#Selecciona todas las columnas de las filas 2 a 4 (inclusive) utilizando loc.

worldData.loc[2:4:]
```

```python
# Utilizando loc, filtra el conjunto de datos para mostrar solo las filas donde la densidad (columna "Birth Rate") es mayor a 30.

maskDensidad = worldData["Birth Rate"] > 30

worldData.loc[maskDensidad].head(5)
```

```python
#Utilizando loc selecciona las columnas "Country" y "Capital/Major City" de las filas donde el porcentaje de tierra agrícola (columna "Fertility Rate") es superior al 3.

maskAgricola = worldData["Fertility Rate"] > 3

worldData.loc[maskAgricola][["Country","Capital/Major City"]].head(5)
```

```python
worldData.iloc[0:4].head()
```

```python
#Selecciona todas las filas de las columnas 0 a 4 (excluyendo la columna 4) utilizando iloc.

worldData.iloc[::,[0,1,2,3]]
```

```python
#Selecciona todas las filas de las columnas 0 a 4 (excluyendo la columna 4) utilizando iloc.

worldData.iloc[::,0:4]
```

```python
#Encuentra el valor en la fila 2 y columna 7 utilizando iloc.

worldData.iloc[1,7]
```

```python
#Selecciona todas las filas para las columnas 0, 2, y 5 utilizando iloc.

worldData.iloc[::,[0,2,5]].head()
```

```python
worldData.head()
```

```python
#Encuentra el valor de "Gasoline Price" para el país cuya abreviatura es "AL" utilizando loc.

maskAbr = worldData["Abbreviation"] == "AL"
worldData.loc[maskAbr,"Gasoline Price"].head()
```
