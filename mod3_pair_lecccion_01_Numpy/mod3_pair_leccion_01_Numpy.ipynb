Creación e Indexación:

```python
# Ejercicio 1: Crea un array 1D con 50 valores aleatorio del 10 al 99. Accede a los elementos en las posiciones múltiplos de 5.

a1 = np.random.randint(10, 99, (50,))
print(a1)

print("-----")

multi = a1[::5]
print(multi)
```

```python
# Ejercicio 2: Crea un array 1D con 30 valores aleatorios del 1 al 20 (ambos incluidos). Utiliza indexación para obtener los elementos en las posiciones impares.


a1_2 = np.random.randint(1, 21, (30,))
print(a1_2)

print("-----")

imp = a1_2[1::2]
print(imp)
```

```python
# Ejercicio 3: Crea un array 2D de forma (4, 6) con números aleatorios decimales (no deben estar comprendidos solo entre 0 y 1). Accede a la última fila de la matriz. Los decimales deben estar redondeados a 2 decimales.

d2 = np.random.uniform(1, 5, (4, 6))  # numpy.random.randint(low, high, size, dtype)
print(d2)

print("-----")

print(d2[-1])

print("-----")

redondeo = np.round(d2, 2) 
print(redondeo)
```

Funciones Aritméticas y Estadísticas:

```python
# Ejercicio 1: Crea un array de tres 3D con elementos entre el 0 y 1 de forma (2, 3, 5) cuyos valores solo tengan 1 decimal.

d3 = np.round(np.random.rand(2, 3, 5), 1)
print(d3)
```

```python
# Ejercicio 2: Crea un array vacío con la misma forma y dimensiones que el array creado anteriormente. Vamos a reemplazar los valores de la matriz vacía que nos acabamos de crear pero basándonos en los valores del primer array bajo las siguientes condiciones:

d3_2 = np.empty((2, 3, 5), dtype = "U1") # Para crear un espacio vacío nuevo
print(d3_2)
```

```python
# Si el valor del array es igual que la media, se reemplaza por "A".

media = np.round(np.mean(d3), 1)
print("media:", media)

print("-----")

d3_2[d3 == media] = "A"
print(d3_2)

print("-----")

# Si el valor del array es mayor que la media, se reemplaza por "B".

d3_2[d3 > media] = "B"
print(d3_2)

print("-----")

# Si el valor del array es menor que la media, se reemplaza por "C".

d3_2[d3 < media] = "C"
print(d3_2)

print("-----")

# Si el valor es igual que el máximo, se reemplaza por "D".

max = np.max(d3)
print("máximo:", max)

print("-----")

d3_2[d3 == max] = "D"
print(d3_2)

print("-----")

# Si el valor es igual que el mínimo, se reemplaza por "E".

min = np.min(d3)
print("mínimo:", min)

print("-----")

d3_2[d3 == min] = "E"
print(d3_2)
```

```python
# Ejercicio 1: Crea un array 3D de forma (2, 3, 5) y 30 elementos con números enteros entre 0 y 100 (ambos incluidos). 
# Reemplaza los elementos en posiciones múltiplos de 3 por 0.

d3_3 = np.random.randint(0, 101, (2, 3, 5))

print(d3_3)
```

```python
mask = (d3_3 % 3 == 0)

result = np.where(mask, 0, d3_3)

print(result)
```

```python
# Ejercicio 2: Crea un array 3D de forma (2, 3, 4) con valores aleatorios entre 1 y 100 (ambos inclusive). 
# # Utiliza reshape para convertirlo en un array 2D. Deberás pensar en la forma que es correcta para hacer este cambio.

d3_4 = np.random.randint(0, 101, (2, 3, 4))

print(d3_4)
```

```python
d3_4_vs2 = d3_4.reshape(3, 8) # El total de valores del shape de 2d debe ser el mismo total de valores del shape de 3d

print(d3_4_vs2)
```

```python
# Ejercicio 3: Crea un array 3D de forma (3, 4, 5) con valores enteros del 1 al 61 (sin incluir).
# Utiliza reshape para convertirlo en un array 2D.


d3_5 = np.random.randint(1, 61, (3, 4, 5)) # Original -> A través del índice: (3 = 0, 4 = 1, 5 = 2)

print(d3_5)
```

```python
d3_5_vs2 = d3_5.reshape(6, 10) # El total de valores del shape de 2d debe ser el mismo total de valores del shape de 3d

print(d3_5_vs2)
```

```python
# Ejercicio 4: Utiliza el array 3D del ejercicio anterior de reshape. 
# Utiliza transpose para permutar los ejes y obtener un nuevo array con forma (5, 4, 3). 
# Recordatorio_original = (3 = 0, 4 = 1, 5 = 2)

d3_5_trans = d3_5.transpose(2, 1, 0)

print(d3_5_trans)
```

```python
# Ejercicio 5: Utiliza el array 3D del ejercicio anterior 3 de reshape. 
# Utiliza transpose para cambiar el orden de los ejes y obtener un nuevo array con forma (3, 5, 4).
# Recordatorio_original = (3 = 0, 4 = 1, 5 = 2)

d3_5_trans2 = d3_5.transpose(0, 2, 1)

print(d3_5_trans2)
```
