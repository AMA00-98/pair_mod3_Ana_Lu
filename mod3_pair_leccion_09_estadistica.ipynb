## Ejercicio 4: Países más y menos ricos --> 'gdp'

```python
# Caso A: Calcula la mediana 
print(f"Gdp:\nMediana: {df["gdp"].median().round(2)}\n")
```

```python
#Crea dos grupos de países:
    #grupo_bajo_gdp: países con gdp por debajo (o igual) de la mediana.

    #grupo_alto_gdp: países con gdp por encima de la mediana.

#pd.cut(df["gdp"], bins=[(df["gdp"]<df["gdp"].median()),(df["gdp"]>df["gdp"].median())], labels=["Bajo", "Alto"])


grupo_bajo_gdp = df[df["gdp"]<df["gdp"].median()]

grupo_alto_gdp = df[df["gdp"]>df["gdp"].median()]
```

```python
#Para cada grupo, calcula la media y la mediana de: life_expectancy, infant_mortality
columnas = ["life_expectancy", "infant_mortality"]
            
print(f"grupo_bajo_gdp - life_expectancy:\nMedia: {grupo_bajo_gdp["life_expectancy"].mean().round(2)}\nMediana: {grupo_bajo_gdp["life_expectancy"].median().round(2)}\n")

print(f"grupo_bajo_gdp - infant_mortality:\nMedia: {grupo_bajo_gdp["infant_mortality"].mean().round(2)}\nMediana: {grupo_bajo_gdp["infant_mortality"].median().round(2)}\n\n")

print(f"grupo_alto_gdp - life_expectancy:\nMedia: {grupo_alto_gdp["life_expectancy"].mean().round(2)}\nMediana: {grupo_alto_gdp["life_expectancy"].median().round(2)}\n")

print(f"grupo_alto_gdp - infant_mortality:\nMedia: {grupo_alto_gdp["infant_mortality"].mean().round(2)}\nMediana: {grupo_alto_gdp["infant_mortality"].median().round(2)}\n")
```

```python
# Interpreta:

    # ¿En qué grupo la esperanza de vida es mayor, de media? En el grupo alto, la esperanza de vida es mayor con 75,5 años en lugar de 69 años correspondientes al grupo bajo.

    # ¿En qué grupo la mortalidad infantil es más alta? En este caso, en el grupo bajo con un 28.82% frente al 14.02% del grupo alto.

    # En lenguaje de políticas públicas, ¿qué conclusión preliminar podrías sacar de esta comparación simple? 
        #Esta brecha puede sugerir un cambio de tecnología avanzada, medicina, infraestructuras debido a la economía.
```

```python
#1. Calcula:
    #Media
    #Mediana
    #Desviación estándar
    #Percentiles 25, 50 y 75.

print(f"population_labor_force_participation:\nMedia: {df["population_labor_force_participation"].mean().round(2)}\nMediana: {df["population_labor_force_participation"].median().round(2)}\nDesviación: {df["population_labor_force_participation"].std().round(2)}\n\n")

#usamos nanpercentile porque contiene nulos, en lugar de percentile

q1 = np.nanpercentile(df["population_labor_force_participation"], 25).round(2)
q2 = np.nanpercentile(df["population_labor_force_participation"], 50)
q3 = np.nanpercentile(df["population_labor_force_participation"], 75)

print(f"population_labor_force_participation:\nQ1 = {q1}\nQ2 = {q2}\nQ3 = {q3}") 
```
