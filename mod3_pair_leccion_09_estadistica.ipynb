## Ejercicio 1: Calcular

```python
columnas = ["population", "density", "life_expectancy", "fertility_rate", "gdp"]
```

```python
# Caso A: Media

for dato in columnas:
    print(f"La Media de '{dato}' es: {round(np.mean(df[dato]), 2)}")
```

```python
# Caso B: Mediana

for dato in columnas:
    print(f"La Mediana de '{dato}' es: {round(np.nanmedian(df[dato]), 2)}")
```

```python
# Caso C: Desviación estándar

for dato in columnas:
    print(f"La Desviación estándar de '{dato}' es: {round(np.std(df[dato]), 2)}")
```

```python
# Caso D: Mínimo

for dato in columnas:
    print(f"El Mínimo de '{dato}' es: {round(np.min(df[dato]), 2)}")
```

```python
# Caso E: Máximo

for dato in columnas:
    print(f"El Máximo de '{dato}' es: {round(np.max(df[dato]), 2)}")
```
