## Ejercicios 'apply'

```python
# Ejercicio 1: Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. 
# Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

df_wdata[["land_area", "armed_forces_size", "co2-emissions", "gdp", "population", "urban_population", "latitude", "longitude", "density", "cpi"]].head()

# land_area + gdp + population + urban_population -> int
# armed_forces_size + co2-emissions + cpi -> float
```

```python
# info original
df_wdata.info()
```

```python
def cambio_int(nombre):
    try:
        return int(nombre.replace(",","").replace(".",""))
    
    except:
        return nombre 
    
def cambio_float(nombre):
    try:
        return float(nombre.replace(",","").replace(".",""))
    
    except:
        return nombre 
```

```python
# Pruebas
print(cambio_int("652,230"))
print(cambio_float("323,000"))
```

```python
col_int = ["land_area", "gdp", "population", "urban_population"]
```

```python
# land_area + gdp + population + urban_population -> int

for columna in col_int:

    df_wdata[columna] =  df_wdata[columna].apply(cambio_int)
```

```python
df_wdata.info()
```

```python
col_float = ["armed_forces_size", "co2-emissions", "cpi"]
```

```python
# armed_forces_size + co2-emissions + cpi -> float

for columna in col_float:

    df_wdata[columna] =  df_wdata[columna].apply(cambio_float)
```
