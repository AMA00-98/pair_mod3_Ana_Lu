## Después de examinar los datos, vemos las siguientes coincidencias:

- Country vs country (1,2)
- Capital/Major City vs Largest city (1,2)

```python
#Utiliza el método de unión de Pandas que consideres más adecuado para combinar los datos de ambos años en un solo DataFrame.
```

```python
df_wd2023_1.merge(df_wd2023_2,left_on=["Country","Capital/Major City"],right_on=["country","Largest city"]).head(2)
```

```python
df_wd2023 = df_wd2023_1.merge(df_wd2023_2,left_on=["Country","Capital/Major City"],right_on=["country","Largest city"]) #guardamos cambios en la variable
```

```python
df_wd2023.head(2)
```

#Explica por qué elegiste ese método de unión y cómo se llevaron a cabo los pasos anteriores.

Hemos elegido merge al tener columnas comunes entre ambos dataframes, aunque con nombres diferentes. Hemos indicado que la unión se realice a través de las columnas "Country" y "Capital/Major City" del primer dataframe y "country" y "Largest city" del segundo dataframe. De esta forma, se han combinado los datos de ambos años en un solo DataFrame, manteniendo solo las filas que tienen coincidencias en ambas tablas.

## Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

    - No tengan espacios. X
    - Estén en minúscula. X
    - No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
    - Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
    - Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
nuevas_columnas = {columna: columna.lower().replace(" ","") for columna in df_wd2023.columns} #vamos a iterar por todas las columnas del DataFrame y a cada una de ellas la pondremos en minúsula y le aplicaremos un replace para quitar los puntos. 
  
#todo minuscula y sin espacios
#comprobamos que hemos creado el diccionario correctamente
nuevas_columnas

# una vez que tenemos el diccionario vamos a aplicar el método rename
df_wd2023.rename(columns = nuevas_columnas, inplace = True)

df_wd2023.head(2)
```

```python
#PRIMERA PRUEBA. iteramos en busqueda de la solución.
for c in df_wd2023.columns:
        print (c.split("(")[0].strip().replace(":","").lower()) 
```

```python
#split por la parte de apertura y nos quedamos con la parte izquierda.

nuevas_columnas2 = {c: c.split("(")[0].strip().replace(":","").lower() for c in df_wd2023.columns} #strip quita el \n

nuevas_columnas2

# una vez que tenemos el diccionario vamos a aplicar el método rename
df_wd2023.rename(columns = nuevas_columnas2, inplace = True)

df_wd2023.head(2)
```
