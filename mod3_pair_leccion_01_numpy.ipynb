Creación e Indexación:

```python
# Ejercicio 1: Crea un array 1D con 50 valores aleatorio del 10 al 99. Accede a los elementos en las posiciones múltiplos de 5.

a1 = np.random.randint(10, 99, (50,))
print(a1)

print("-----")

multi = a1[::5]
print(multi)
```

```python
# Ejercicio 2: Crea un array 1D con 30 valores aleatorios del 1 al 20 (ambos incluidos). Utiliza indexación para obtener los elementos en las posiciones impares.


a1_2 = np.random.randint(1, 21, (30,))
print(a1_2)

print("-----")

imp = a1_2[1::2]
print(imp)
```

```python
# Ejercicio 3: Crea un array 2D de forma (4, 6) con números aleatorios decimales (no deben estar comprendidos solo entre 0 y 1). Accede a la última fila de la matriz. Los decimales deben estar redondeados a 2 decimales.

d2 = np.random.uniform(1, 5, (4, 6))  # numpy.random.randint(low, high, size, dtype)
print(d2)

print("-----")

print(d2[-1])

print("-----")

redondeo = np.round(d2, 2) 
print(redondeo)
```

Funciones Aritméticas y Estadísticas:

```python
# Ejercicio 1: Crea un array de tres 3D con elementos entre el 0 y 1 de forma (2, 3, 5) cuyos valores solo tengan 1 decimal.

d3 = np.round(np.random.rand(2, 3, 5), 1)
print(d3)
```

```python
# Ejercicio 2: Crea un array vacío con la misma forma y dimensiones que el array creado anteriormente. Vamos a reemplazar los valores de la matriz vacía que nos acabamos de crear pero basándonos en los valores del primer array bajo las siguientes condiciones:

d3_2 = np.empty((2, 3, 5), dtype = "U1") # Para crear un espacio vacío nuevo
print(d3_2)
```

```python
# Si el valor del array es igual que la media, se reemplaza por "A".

media = np.round(np.mean(d3), 1)
print("media:", media)

print("-----")

d3_2[d3 == media] = "A"
print(d3_2)

print("-----")

# Si el valor del array es mayor que la media, se reemplaza por "B".

d3_2[d3 > media] = "B"
print(d3_2)

print("-----")

# Si el valor del array es menor que la media, se reemplaza por "C".

d3_2[d3 < media] = "C"
print(d3_2)

print("-----")

# Si el valor es igual que el máximo, se reemplaza por "D".

max = np.max(d3)
print("máximo:", max)

print("-----")

d3_2[d3 == max] = "D"
print(d3_2)

print("-----")

# Si el valor es igual que el mínimo, se reemplaza por "E".

min = np.min(d3)
print("mínimo:", min)

print("-----")

d3_2[d3 == min] = "E"
print(d3_2)
```
